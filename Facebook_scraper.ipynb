```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import re

# Path to the ChromeDriver executable
driver_path = 'D:\\hp5cd\\New folder\\Web_Scraping\\chromedriver-win64\\chromedriver.exe'

# Set up Chrome options
options = Options()
options.add_argument("--disable-gpu")

# Set up the Service object
service = Service(executable_path=driver_path)

# Initialize the WebDriver
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()

def get_email_from_profile(driver, profile_url):
    driver.get(profile_url)
    try:
        # Attempt to find the email element
        email_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "span.x193iq5w.xeuugli.x13faqbe.x1vvkbs.x1xmvt09.x1lliihq.x1s928wv.xhkezso.x1gmr53x.x1cpjm7i.x1fgarty.x1943h6x.xudqn12.x3x7a5m.x6prxxf.xvq8zen.xo1l8bm.xzsf02u.x1yc453h"))
        )                                                          
        # Return the email text if found
        email = email_element.text if email_element else '-'
    except Exception as e:
        # Handle cases where the email element is not found
        print(f"Error finding email: {e}")
        email = '-'
    return email

def process_member(driver, member_element):
    try:
        href = member_element.get_attribute('href')
        print(f"Debug: href = {href}")  # Print the href for debugging

        # Define a regex pattern to extract user IDs
        pattern = r'/user/(\d+)/'
        match = re.search(pattern, href)

        if match:
            user_id = match.group(1)
            profile_url = f"https://www.facebook.com/profile.php?id={user_id}&sk=about_contact_and_basic_info"
            email = get_email_from_profile(driver, profile_url)
            return (user_id, email)
        else:
            print(f"Skipping invalid user link: {href}")
            return None
    except Exception as e:
        print(f"Error processing element: {e}")
        return None

def extract_ids_and_scrape(driver, group_members_url):
    driver.get(group_members_url)
    time.sleep(30)  # Wait for the page to load initially

    results = []

    while True:
        try:
            # Re-fetch member elements
            member_elements = driver.find_elements(By.CSS_SELECTOR, "a.x1i10hfl.xjbqb8w.x1ejq31n.xd10rxx.x1sy0etr.x17r0tee.x972fbf.xcfux6l.x1qhh985.xm0m39n.x9f619.x1ypdohk.xt0psk2.xe8uvvx.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x16tdsg8.x1hl2dhg.xggy1nq.x1a2a7pz.x1sur9pj.xkrqix3.xzsf02u.x1s688f")

            if not member_elements:
                print("No more member elements found.")
                break

            # Process each member element and handle exceptions
            for member_element in member_elements:
                result = process_member(driver, member_element)
                if result:
                    results.append(result)
                time.sleep(2)  # Wait before processing the next URL
            
            # Scroll to load more elements
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(5)  # Wait for the page to load more content
            
        except Exception as e:
            print(f"Error during extraction: {e}")
            break

    return results

group_members_url = 'https://www.facebook.com/groups/362424800118725/members'
results = extract_ids_and_scrape(driver, group_members_url)

# Save results to a CSV file
with open('scraped_emails.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['User ID', 'Email'])
    writer.writerows(results)

driver.quit()
```
